# Анализ бизнес-показателей

Проект в рамках стажировки в отдел маркетинговой аналитики Яндекс.Афиши. Первое задание: помочь маркетологам снизить расходы — отказаться от невыгодных источников трафика и перераспределить бюджет

Есть данные Яндекс.Афиши с июня 2017 по конец мая 2018 года:


- лог сервера с данными о посещениях сайта Яндекс.Афиши,
- выгрузка всех заказов за этот период,
- статистика рекламных расходов.


Необходимо предстоит изучить:


- как клиенты пользуются сервисом,
- когда делают первые покупки на сайте,
- сколько денег приносит компании каждый клиент,
- когда расходы на привлечение клиента окупаются.


Выручка измеряется в условных единицах — у.е.

```python
# Для формирования метрик вводим дополнительные поля
visits['year'] = visits['start_ts'].astype('datetime64[Y]')
visits['month'] = visits['start_ts'].astype('datetime64[M]')
visits['week'] = visits['start_ts'].astype('datetime64[W]')
visits['date'] = visits['start_ts'].dt.date
visits.head()
```

Вычисляем количество уникальных пользователей в различные периоды

```python
dau = visits.groupby('date').agg({'uid':'nunique'})
dau_mean = dau.mean()
```

```python
wau = visits.groupby('week').agg({'uid':'nunique'})
wau_mean = wau.mean()
```

```python
print('Значение WAU в среднем составляет {:.0f} человек'.format(wau_mean[0]))
```

```python
mau = visits.groupby('month').agg({'uid':'nunique'})
mau_mean = mau.mean()
```

```python
print('Значение MAU в среднем составляет {:.0f} человек'.format(mau_mean[0]))
```

```python
def plot_au(x,y,z):
    plt.plot(x,label=y)
    plt.grid(True)
    plt.hlines(z,'2017-06-01','2018-05-31',color='red',label='Среднее')
    plt.title('Изменение метрики во времени')
    plt.xlabel('Дата')
    plt.ylabel('Значение, ед.')
    plt.legend()
    plt.show()
```

```python
plot_au(dau,'DAU',dau_mean)
import warnings
warnings.filterwarnings('ignore')
```

```python
plot_au(wau,'WAU',wau_mean)
```

```python
plot_au(mau,'MAU',mau_mean)
```

#### Retention rate

```python
#  формируем когорты пользователей по первому посещению сайта
first_visit = visits.groupby('uid')['date'].min()
first_visit.name = 'first_visit_date'
visits_min = visits.join(first_visit, on='uid')
visits_min['first_visit_month'] = visits_min['first_visit_date'].astype('datetime64[M]')
```

```python
visits_min.head()
```

```python
#  распределяем действия пользователей по периодам
visits_min['cohort_lifetime'] = (visits_min['month'] - visits_min['first_visit_month'])
visits_min['cohort_lifetime'] = visits_min['cohort_lifetime'] / np.timedelta64(1, 'M')
visits_min['cohort_lifetime'] = visits_min['cohort_lifetime'].round().astype('int')
```

```python
cohort_visits = visits_min.groupby(['first_visit_month','cohort_lifetime']
                              ).agg({'uid':'nunique'}).reset_index()
cohort_visits
```

```python
#  вычисляем численность когорт
users_count = cohort_visits[cohort_visits['cohort_lifetime'] == 0][['first_visit_month', 'uid']]
```

```python
users_count = users_count.rename(columns={'uid':'cohort_users'})
users_count
```

```python
cohort_visits = cohort_visits.merge(users_count, on = 'first_visit_month') 
cohort_visits['retention'] = cohort_visits['uid'] / cohort_visits['cohort_users']
cohort_visits['first_visit_month'] = cohort_visits['first_visit_month'].dt.strftime('%Y-%m-%d')
```

```python
#  формируем таблицу изменения показателя Retention rate в различых когортах
retention_pivot = cohort_visits.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum')
retention_pivot.fillna('')
```

```python
plt.figure(figsize=(12,8))
plt.title('Изменение метрики во времени (Retention rate)')
ax = plt.axes()
sns.heatmap(retention_pivot,
            annot=True,
            fmt='.1%',
            linewidths=1,
            linecolor='gray',
            vmax=0.1,
            ax = ax
           )
ax.set_ylabel('Когорты')
ax.set_xlabel('Время жизни когорт, мес.')
plt.show()
```

```python
retention_mean = retention_pivot[1].mean()
print('Значение Retention rate на второй месяц в среднем по когортам составляет {:.1%}'.format(
    retention_mean))
```

```python
order_count = orders.join(min_purchase,on='uid')
order_count['order_time'] = order_count['buy_ts'].astype('datetime64[M]')
order_count['min_purchase_datetime'] = order_count['min_purchase_datetime'].astype('datetime64[M]')
order_count['delta_time'] = ((order_count['order_time'] - order_count['min_purchase_datetime']
                             ) / np.timedelta64(1,'M')).round().astype('int')
order_count_delta = order_count[order_count['delta_time'] < 7]
order_count_delta = order_count_delta.groupby('uid').agg({'order_time':'count'})
orders_of_six_month = order_count_delta.mean()
print('Среднее количество покупок на одного пользователя за 6 месяцев составляет {:.1f}'.format(orders_of_six_month[0]))
```

#### Изменение показателя LTV на покупателя

```python
#  строим когорты
cohort_sizes = order_count.groupby('min_purchase_datetime').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes
```

```python
cohorts = order_count.groupby(['min_purchase_datetime', 'order_time']).agg({'revenue': 'sum'}).reset_index()
cohorts.columns = ['first_order_month', 'order_time', 'revenue']
```

```python
cohorts.head(10)
```

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report
```

```python
margin_rate = 1 #  эквивалентно маржинальности 100%
report['gp'] = report['revenue'] * margin_rate #  валовый доход
report['age'] = ((report['order_time'] - report['first_order_month']
                ) / np.timedelta64(1, 'M')).round().astype('int')
report['first_order_month'] = report['first_order_month'].dt.strftime('%Y-%m-%d')
report.head()
```

```python
report['ltv'] = report['gp'] / report['n_buyers']
report_ltv = report.pivot_table(
    index='first_order_month',
    columns='age',
    values='ltv',
    aggfunc='sum'
).cumsum(axis=1)
report_ltv.fillna('')
```

```python
plt.figure(figsize=(12,8))
plt.title('Изменение метрики во времени (LTV)')
ax = plt.axes()
sns.heatmap(report_ltv,
            annot=True,
            fmt='.2g',
            linewidths=1,
            linecolor='gray',
            ax = ax
           )
ax.set_ylabel('Когорты')
ax.set_xlabel('Время жизни когорт, мес.')
plt.show()
```

```python
ltv_mean = report_ltv[5].mean()
```

```python
print('Средний LTV по когортам за 6 месяцев составил {:.2}'.format(ltv_mean))
```

Сентябрьская когорта - самая успешная по привлеченному доходу

```python
visiters = visits.sort_values(by = ['uid', 'start_ts']).groupby('uid').agg({'source_id': 'first', 'month':'first'})
customers = orders.join(visiters,on='uid')
customers_by_source = customers.groupby(['source_id','month']).agg({'uid':'nunique'}).reset_index()
customers_by_source.columns = ['source_id','month','users_by_source']
customers_by_source
```

```python
costs['dt'] = costs['dt'].astype('datetime64[M]')
costs_to_cac = costs.sort_values(by = ['source_id','dt']).groupby(['source_id','dt']).agg({'costs':'sum'}).reset_index()
costs_to_cac
```

```python
cac_by_source = pd.merge(customers_by_source,
                         costs_to_cac,
                         how='right',
                         left_on=['source_id','month'],
                         right_on=['source_id','dt'])
cac_by_source.head()
```

```python
cac_by_source['cac'] = cac_by_source['costs'] / cac_by_source['users_by_source']
```

```python
cac_by_source_gr = cac_by_source.groupby('source_id').agg({'cac':'mean'})
cac_by_source_gr
```

```python
cac_by_source['month'] = cac_by_source['month'].dt.strftime('%Y-%m-%d')
cac_pivot = cac_by_source.pivot_table(index='month',columns='source_id',values='cac')
```

```python
plt.figure(figsize=(12,10))
plt.title('Расходы CAC по маркетинговым каналам во времени')
ax = plt.axes()
sns.heatmap(cac_pivot,
            annot=True,
            fmt="g",
            linewidths=1,
            linecolor='gray',
            ax = ax
           )
ax.set_ylabel('Период, мес.')
ax.set_xlabel('Маркетинговые каналы')
plt.show()
```

#### Возврат инвестиций в маркетинг (ROMI) по когортам

```python
#  выделяем пользователей по каналу привлечения на сайт
users = visits.sort_values(by = ['uid', 'start_ts']).groupby('uid').agg({'source_id': 'first', 'month':'first'})
users.info()
```

```python
users_by_orders = orders.sort_values(
    by = ['uid', 'buy_ts'])
users_by_orders['order_time'] = users_by_orders['buy_ts'].astype('datetime64[M]')
```

```python
users_first_order = users_by_orders.groupby('uid')['order_time'].min()
users_first_order.name = 'first_order'
users_by_orders = users_by_orders.join(users_first_order,on='uid')
users_by_orders
```

```python
users_by_orders['age'] = ((users_by_orders['order_time'] - users_by_orders['first_order'])/ np.timedelta64(1, 'M')).round().astype('int')
users_by_orders.head()
```

```python
source_users = pd.merge(users_by_orders,users,on='uid')
source_users.head()
```

```python
source_users = source_users.groupby(['source_id','age']).agg({'revenue':'sum'}).reset_index()
source_users
```

```python
source_romi = pd.merge(source_users,costs_to_cac,left_index=True,right_index=True)
source_romi['romi'] = source_romi['revenue']/source_romi['costs']
```

```python
source_romi
```

```python
output = source_romi.pivot_table(
    index='source_id_y', columns='age', values='romi', aggfunc='mean'
)
output.cumsum(axis=1).round(2).fillna('')
```

```python
#   строим график распределения ROMI по когортам в разрезе маркетинговых каналов
plt.figure(figsize=(12,10))
plt.title('Изменение метрики во времени (ROMI)')
ax = plt.axes()
sns.heatmap(output.cumsum(axis=1),
            annot=True,
            fmt=".2f",
            linewidths=1,
            linecolor='gray',
            ax = ax
           )
ax.set_ylabel('Маркетинговые каналы')
ax.set_xlabel('Период жизни когорт, мес.')
plt.show()
```
